```python

from sqlalchemy import create_engine
import pandas as pd

# Define connection parameters
user = "root"
password = "root"
host = "localhost"
port = "3306"
database = "phonepe"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

# Connect to the database
conn = engine.connect()

# Case Study 1:Insurance_Insights Districts Vs Transaction Amount 
query1 = """
 SELECT districts, 
    ROUND(SUM(Transactions_amount)/1000000000000,5) AS Trans_Amount_Trillion
    from top_trans group by districts ORDER BY 
    Trans_Amount_Trillion DESC limit  20;
    
    """
   
# Execute query and load into DataFrame
df8 = pd.read_sql(query1, conn)
df8['query'] = 'query1'
```

```python
df8
```

```python
# Case Study 2:Insurance_Insights State Vs Transaction Count 
query2 = """
SELECT 
    State, 
    ROUND(SUM(Transactions_count)/1000000000,2) AS Transactions_Count_Crore
 FROM 
    top_trans
GROUP BY 
    State
ORDER BY 
    Transactions_Count_Crore DESC
LIMIT 10;
    
    """
      
# Execute query and load into DataFrame
df9 = pd.read_sql(query2, conn)
df9['query'] = 'query2'

```

```python
df9
```

```python
from numpy import concat

# Step 3: Combine all results into one CSV
insu_top_combined_df = pd.concat([df8, df9], ignore_index=True)
insu_top_combined_df.fillna('', inplace=True)
insu_top_combined_df.to_csv("insu_top_case_combined.csv", index=False)


print("case1_combined.csv created successfully!")
```
