```python
from sqlalchemy import create_engine
import pandas as pd

# Define connection parameters
user = "root"
password = "root"
host = "localhost"
port = "3306"
database = "phonepe"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

# Connect to the database
conn = engine.connect()

# Case Study 1: Customer Segmentation - Transaction Type Vs transaction_count
query1 = """
SELECT     Transaction_type,
    ROUND(SUM(Transaction_count) / 10000000, 5) AS Total_count_Crore
FROM 
    aggregated_trans
GROUP BY 
    Transaction_type
ORDER BY 
    Total_count_Crore DESC;
    """
# Execute query and load into DataFrame
df1 = pd.read_sql(query1, conn)
df1['query'] = 'query1'
```

```python
df1
```

```python
### Case Study 2: Customer Segmentation - Transaction_Amount Vs Year

query2 = """
SELECT 
    Year, 
    ROUND(SUM(Transaction_amount) / 1000000000000, 2) AS Total_amount_Trillion
FROM 
    aggregated_trans
GROUP BY 
    Year
ORDER BY 
    SUM(Transaction_amount) DESC;
"""
# Execute query and load into DataFrame
df2 = pd.read_sql(query2, conn)
df2['query'] = 'query2'
```

```python
df2
```

```python
from numpy import concat

# Step 3: Combine all results into one CSV
combined_df = pd.concat([df1, df2], ignore_index=True)
combined_df.fillna('', inplace=True)
combined_df.to_csv("case1_combined.csv", index=False)


print("case1_combined.csv created successfully!")
```
