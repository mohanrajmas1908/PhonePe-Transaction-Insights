```python

from sqlalchemy import create_engine
import pandas as pd

# Define connection parameters
user = "root"
password = "root"
host = "localhost"
port = "3306"
database = "phonepe"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

# Connect to the database
conn = engine.connect()

# Case Study 1: Trend Analysis
query1 = """
SELECT 
    Year,
    Quarter,
    ROUND(SUM(Transaction_amount) / 1000000000, 2) AS Total_Amount_Billion
FROM 
    aggregated_trans
GROUP BY 
    Year, Quarter
ORDER BY 
    Year, Quarter;
"""
   
# Execute query and load into DataFrame
df7 = pd.read_sql(query1, conn)
df7['query'] = 'query1'
```

```python
df7
```

```python
from numpy import concat

# Step 3: Combine all results into one CSV
trend_combined_df = pd.concat([df7], ignore_index=True)
trend_combined_df.fillna('', inplace=True)
trend_combined_df.to_csv("Trend_Ana_case_combined.csv", index=False)


print("Trend_Ana_case_combined.csv created successfully!")
```
