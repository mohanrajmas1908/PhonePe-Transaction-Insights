```python

from sqlalchemy import create_engine
import pandas as pd

# Define connection parameters
user = "root"
password = "root"
host = "localhost"
port = "3306"
database = "phonepe"

# Create SQLAlchemy engine
engine = create_engine(f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}")

# Connect to the database
conn = engine.connect()

# Case Study 1: User_Engagement
query1 = """
  SELECT  a.Brand, 
    CAST(ROUND(SUM(b.AppOpens) / 1000000000	, 0) AS UNSIGNED) AS AppOpens_Billion
FROM aggregated_user a 
JOIN  map_user b ON a.State = b.State 
GROUP BY a.Brand
    """
   
# Execute query and load into DataFrame
df5 = pd.read_sql(query1, conn)
df5['query'] = 'query1'
```

```python
df5
```

```python
# Case Study 2: User_Engagement-Distribution of Registered Users Across Brands
query2 = """
 SELECT  a.Brand, 
     CAST(ROUND(SUM(b.RegisteredUsers) / 1000000000, 0) AS UNSIGNED) AS RegisteredUsers_Billion
    FROM aggregated_user a 
JOIN  map_user b ON a.State = b.State
GROUP BY a.Brand ;
    """
      
# Execute query and load into DataFrame
df6 = pd.read_sql(query2, conn)
df6['query'] = 'query2'

```

```python
df6
```

```python
from numpy import concat

# Step 3: Combine all results into one CSV
user_combined_df = pd.concat([df5, df6], ignore_index=True)
user_combined_df.fillna('', inplace=True)
user_combined_df.to_csv("User_case_combined.csv", index=False)


print("user_combined_df.csv created successfully!")
```
