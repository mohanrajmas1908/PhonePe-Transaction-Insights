```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="C:\\DataScience\\Projects\\pulse\\data\\top\\insurance\\country\\india\\state\\"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quarter':[],'districts':[], 'Trans_count':[], 'Trans_amount':[],
    'pincodes' :[],'Pin_Trans_count':[],'Pin_Trans_amount':[]
     
     
     }

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']["districts"]:
                 name = z["entityName"]
                 count = z["metric"]["count"]
                 amount = z["metric"]["amount"]
            for c in D['data']["pincodes"]:
                 name_1 = c["entityName"]
                 count_1 = z["metric"]["count"]
                 amount_1 = z["metric"]["amount"]
                 clm['districts'].append(name)
                 clm['Trans_count'].append(count)
                 clm['Trans_amount'].append(amount)
                 clm['pincodes'].append(name_1)
                 clm['Pin_Trans_count'].append(count_1)
                 clm['Pin_Trans_amount'].append(amount_1)
                 clm['State'].append(i)
                 clm['Year'].append(j)
                 clm['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Top_insurance=pd.DataFrame(clm)
```

```python
# Remove dupdrop_duplicates
Top_insurance_unique = Top_insurance.drop_duplicates()
```
