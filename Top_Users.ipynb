```python
#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="C:\\DataScience\\Projects\\pulse\\data\\top\\user\\country\\india\\state\\"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quarter':[],'districts':[], 'registeredUsers':[],'pincodes':[],'Pin_registeredUsers':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']["districts"]:
                 name = z["name"]
                 count = z["registeredUsers"]
            for c in D['data']["pincodes"]:
                 name_1 = c["name"]
                 count_1 = c["registeredUsers"]
                 clm['districts'].append(name)
                 clm['registeredUsers'].append(count)
                 clm['pincodes'].append(name_1)
                 clm['Pin_registeredUsers'].append(count_1)
                 clm['State'].append(i)
                 clm['Year'].append(j)
                 clm['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Top_user=pd.DataFrame(clm)
```

```python
# Remove dupdrop_duplicates
Top_user_unique = Top_user.drop_duplicates()
```

```python
#To remove null (missing) values
Top_user_cleaned = Top_user_unique.dropna()
```
