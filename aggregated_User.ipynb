```python

import os
path="C:\\DataScience\\Projects\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Agg_state_list=os.listdir(path)

#Once created the clone of GIT-HUB repository then,
#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path= "C:\\DataScience\\Projects\\pulse\\data\\aggregated\\user\\country\\india\\state\\"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#<------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------>#

#This is to extract the data's to create a dataframe


clm = {
    'State': [], 'Year': [], 'Quarter': [],
    'Brand': [], 'Transaction_count': [], 'Percentage': [],
}


for i in Agg_state_list:
    p_i = os.path.join(path, i)
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = os.path.join(p_i, j)
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = os.path.join(p_j, k)
            try:  
                with open(p_k, 'r') as Data:
                    D = json.load(Data)
                users = D.get("data", {}).get("usersByDevice")
                if users:  # Only loop if it's not None or empty
                    for device in users:
                        clm["State"].append(i)
                        clm["Year"].append(j)
                        clm["Quarter"].append(int(k.strip('.json')))
                        clm["Brand"].append(device["brand"])
                        clm["Transaction_count"].append(device["count"])
                        clm["Percentage"].append(device["percentage"])
                else:
                    print(f"Skipping file {p_k}: usersByDevice is None or empty")
            except Exception as e:
                print(f"Error reading file {p_k}: {e}")

#  Create the DataFrame
Agg_User = pd.DataFrame(clm)
Agg_User
```
